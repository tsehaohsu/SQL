**Example 4** \
We want to send all of our high-value customers a special VIP gift. We're defining high-value
customers as those who've made at least 1 order with a total value (not including the discount) equal to
$10,000 or more. We only want to consider orders made in the year 2016.


```python
%sql SELECT * FROM customers LIMIT 2;
```

```python
%sql SELECT * FROM orders LIMIT 2;
```

```python
%sql SELECT * FROM order_details LIMIT 2;
```

```python
%sql SELECT company_name AS customer, ROUND(SUM(c.unit_price*c.quantity)::DECIMAL,2) AS order_total  \
     FROM customers a INNER JOIN orders b on a.customer_id = b.customer_id \
                    INNER JOIN order_details c on b.order_id=c.order_id \
     WHERE TO_CHAR(b.order_date, 'YYYY')='1996' \
     GROUP BY 1 \
     HAVING SUM(c.unit_price*c.quantity) > 10000 \
     ORDER BY 2 DESC \
     LIMIT 5;
```

**Example 5** \
At the end of the month, salespeople are likely to try much harder to get orders, to meet their month-end
quotas. Show all orders made on the last day of the month. Order by `Employee_ID` and `Order_ID`.

```python
%sql SELECT employee_id, order_id, order_date \
     FROM orders \
     WHERE order_date=TO_DATE(TO_CHAR(order_date,'YYYY') || '-12-31','YYYY-MM-DD') OR \
           order_date=TO_DATE(TO_CHAR(order_date,'YYYY')||'-'|| \
     CASE TO_CHAR(order_date,'MM') \
            WHEN '01' THEN '02' \
            WHEN '02' THEN '03' \
            WHEN '03' THEN '04' \
            WHEN '04' THEN '05' \
            WHEN '05' THEN '06' \
            WHEN '06' THEN '07' \
            WHEN '07' THEN '08' \
            WHEN '08' THEN '09' \
            WHEN '09' THEN '10' \
            WHEN '10' THEN '11' \
            WHEN '11' THEN '12' \
            WHEN '12' THEN '01' \
            end ||'-01','YYYY-MM-DD')-1 \
     ORDER BY employee_id, order_id \
     LIMIT 10;
```

Explanation:
1. Extract the year of `order_date` and convert to string.
2. Extract the month of `order_date` and convert to string.
3. Add one monthe to current month.
4. Concatenate the year and month, set the day as the first day of the month.
5. Covnert the string to date data type.
6. Since the new date is always the first day of next month, then minus one day become the last day of current month.
7. Special casse is the last day of year (Dec. 31st)

**Example 7** \
The Northwind mobile app developers would now like to just get a random assortment of orders for beta
testing on their app. Show a random set of 2% of all orders.

```python
%sql WITH x as (SELECT MAX(order_id) max_id FROM orders) \
     SELECT order_id+FLOOR(RANDOM()*100) random_order_id \
     FROM orders  \
     WHERE order_id+FLOOR(RANDOM()*100) <=(select max_id from x) \
     LIMIT (SELECT FLOOR(COUNT(*)*0.02) FROM orders);
```

```python
# Another way
%sql WITH x as (SELECT MAX(order_id) max_id FROM orders) \
     SELECT FLOOR(RANDOM()*(SELECT max_id from x)) random_order_id \
     FROM orders \
     LIMIT (SELECT FLOOR(COUNT(*)*0.02) FROM orders);
```

Explanation:
1. With is a temporary table called x. In this case, it contain one cell which has max_id.
1. Use random() function to generate fractional numbers. The two random() in this code share the same value.
2. Use floor() to convert the fractional number to integer.
3. The random number can not over maximum number of id.
4. Limit the number of data to 2% of all orders.

**Example 8** \
Janet Leverling, one of the salespeople, has come to you with a request. She thinks that she accidentally
double-entered a line item on an order, with a different `Product_ID`, but the same quantity. She
remembers that the quantity was 60 or more. Show all the Order_IDs with line items that match this, in
order of `Order_ID`.

```python
%sql SELECT order_id, quantity, COUNT(*) AS cnt \
     FROM order_details \
     WHERE quantity >= 60 \
     GROUP BY order_id, quantity \
     HAVING COUNT(*) > 1;
```

Based on the previous question, we now want to show details of the order, for orders that match the
above criteria.

```python
%sql SELECT b.order_id, c.quantity \
     FROM employees a \
     INNER JOIN orders b ON a.employee_id=b.employee_id \
     INNER JOIN order_details c ON b.order_id=c.order_id \
     WHERE a.first_name='Janet' AND a.last_name='Leverling' AND c.quantity>60 \
     ORDER BY b.order_id;
```

```python
%sql SELECT b.order_id, a.quantity, c.first_name || ' ' || c.last_name AS employee_name, COUNT(*) AS cnt \
     FROM order_details a \
     INNER JOIN orders b ON a.order_id = b.order_id \
     INNER JOIN employees c on b.employee_id = c.employee_id \
     WHERE quantity >= 60 AND c.first_name='Janet' AND c.last_name='Leverling' \
     GROUP BY 1,2,3 \
     HAVING COUNT(*) > 1;
```

**Example 9** \
Some salespeople have more orders arriving late than others. Maybe they're not following up on the order
process, and need more training. Which salespeople have the most orders arriving late?

```python
%sql SELECT employees.employee_ID ,last_name ,COUNT(*) total_late_orders \
     FROM orders \
     INNER JOIN employees ON employees.employee_id = orders.employee_id \
     WHERE required_date <= shipped_date \
     GROUP BY employees.employee_id ,employees.last_name \
     ORDER BY total_late_orders DESC;
```

Andrew, the VP of sales, hAS been doing some more thinking about the problem of late orders.
He realizes that just looking at the number of orders arriving late for each salesperson isn't a good idea. 
It needs to be compared against the total number of orders per salesperson.

```python
%sql WITH LateOrders AS ( \
        SELECT employee_id ,COUNT(*) totallateOrders \
        FROM Orders\
        WHERE required_date <= shipped_date \
        GROUP BY employee_id \
        ), AllOrders AS ( \
        SELECT employee_id ,COUNT(*) totalorders \
        FROM orders \
        GROUP BY employee_id \
        ) \
     SELECT employees.employee_id ,last_Name ,AllOrders.totalOrders allorders ,LateOrders.totallateOrders lateorders \
     FROM employees \
     INNER JOIN AllOrders  ON AllOrders.employee_id = employees.employee_id \
     INNER JOIN LateOrders ON lateOrders.employee_id = employees.employee_id;
```

Continuing on the answer for above query, let's fix the results if anyone doesn't have late orders. He/She should have a 0
instead of a Null in Late_Orders.

```python
%sql WITH LateOrders AS ( \
        SELECT employee_id ,COUNT(*) totallateOrders \
        FROM Orders\
        WHERE required_date <= shipped_date \
        GROUP BY employee_id \
        ), AllOrders AS ( \
        SELECT employee_id ,COUNT(*) totalorders \
        FROM orders \
        GROUP BY employee_id \
        ) \
    SELECT employees.employee_id ,last_name ,AllOrders.TotalOrders AllOrders \
            ,COALESCE(LateOrders.totallateorders,0) totalLateOrders\
     FROM employees \
     INNER JOIN AllOrders  ON AllOrders.employee_id = employees.employee_id \
     INNER JOIN LateOrders ON lateOrders.employee_id = employees.employee_id;
```

Explanation
1. COALESCE(NULL,0) to replacee NULL to 0

Now we want to get the percentage of late orders over total orders

```python
%sql WITH x AS (SELECT employee_id, count(*) late \
                FROM orders \
                WHERE shipped_date > required_date \
                GROUP BY 1) \
     SELECT x.employee_id, x.late::DECIMAL/COUNT(*)*100 percentage \
     FROM orders \
     INNER JOIN x ON x.employee_id=orders.employee_id \
     GROUP BY x.employee_id, x.late \
     ORDER BY 2 DESC;
```

**Example 10** \
Andrew Fuller, the VP of sales at Northwind, would like to do a sales campaign for existing customers. He'd like to categorize customers into groups, based on how much they ordered in 1996. Then, depending on which group the customer is in, he will target the customer with different sales materials.

The customer grouping categories are 0 to 1,000, 1,000 to 5,000, 5,000 to 10,000, and over 10,000.


```python
%sql WITH x AS (SELECT a.customer_id, SUM(b.unit_price*b.quantity) total \
                FROM orders a \
                INNER JOIN order_details b ON a.order_id=b.order_id \
                WHERE TO_CHAR(a.order_date,'YYYY')='1996' \
                GROUP BY 1) \
    (SELECT customer_id, 'bronz' customer_level FROM x WHERE ROUND(total) BETWEEN 0 AND 1000 \
     UNION \
     SELECT customer_id, 'silver' customer_level FROM x WHERE ROUND(total) BETWEEN 1000 and 5000 \
     UNION \
     SELECT customer_id, 'gold' customer_level FROM x WHERE ROUND(total) BETWEEN 5000 and 10000 \
     UNION \
     SELECT customer_id, 'platinum' customer_level FROM x WHERE ROUND(total) > 10000 \
    );
```

Explanation
1. UNION is usded for concatenate the rows
